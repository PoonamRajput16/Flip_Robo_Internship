### 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and make data frame.


```python
def IMDB_top_100(url):
    page_2 = requests.get(url)
    soup_2 = BeautifulSoup(page_2.content, 'html.parser')
    
    Movie_Title = []
    Movie_name = soup_2.select("td.titleColumn a")
    for i in Movie_name:
        Movie_Title.append(i.get_text())
    
    Year_Of_Release = []
    movie_year = soup_2.find_all('span',class_='secondaryInfo')
    for i in movie_year:
        Year_Of_Release.append(i.get_text().replace('\n','').replace('(','').replace(')',''))
    
    Rating = []
    movie_rating = soup_2.find_all('td',class_="ratingColumn imdbRating")
    for i in movie_rating:
        Rating.append(i.get_text().replace('\n',''))
    
    IMDB_top_100_movies = pd.DataFrame()

    IMDB_top_100_movies['Movie_Title'] = Movie_Title[0:100]
    IMDB_top_100_movies['Year_Of_Release'] = Year_Of_Release[0:100]
    IMDB_top_100_movies['Rating'] = Rating[0:100]

    print('\n\t\tIMDB’s Top rated 100')
    print('-'*60)
    return(IMDB_top_100_movies)
```

```python
IMDB_top_100('https://www.imdb.com/chart/top/?sort=ir,desc&mode=simple&page=1')
```

### 4. Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’


```python
def Book(URL):
    resource_4 = requests.get(URL)
    soup_4 = BeautifulSoup(resource_4.text, 'html.parser')
    
    Book_title = []
    Title = soup_4.select('h4.italic a')
    for i in Title:
        Book_title.append(i.get_text().replace('★',''))
    
    Author_Name = []
    Author = soup_4.find_all('p',class_='sans bold')
    for i in Author:
        Author_Name.append(i.get_text().replace('\n',''))
    
    Book_genre = []
    Genre = soup_4.find_all('p',class_='genre-links hidden-phone')
    for i in Genre:
        Book_genre.append(i.get_text().strip().replace('\n',''))
    
    Book_review = []
    review = soup_4.find_all('p',class_='excerpt')
    for i in review:
        Book_review.append(i.text.replace('\n',''))
    
    Book_Details = pd.DataFrame({})
    
    Book_Details['Book_title'] = Book_title
    Book_Details['Author_Name'] = Author_Name
    Book_Details['Book_genre'] = Book_genre
    Book_Details['Book_review'] = Book_review
    return(Book_Details)
```

```python
Book('https://bookpage.com/reviews')
```

## 5. Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:

   i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
def Men_Team(url):
    resource_M1 = requests.get(url)
    soup_M1 = BeautifulSoup(resource_M1.content,'html.parser')
    
    M_Team = []
    team = soup_M1.find_all('span',class_='u-hide-phablet')
    for i in team:
        M_Team.append(i.text)
    
    M_Matches = []
    top_match = soup_M1.find_all('td',class_='rankings-block__banner--matches')
    for i in top_match:
        M_Matches.append(i.text)
    match = soup_M1.find_all('td',class_='table-body__cell u-center-text')
    for i in range(0,38,2):
        M_Matches.append(match[i].text)
    
    M_Points = []
    point_top=soup_M1.find_all('td',class_='rankings-block__banner--points')
    for i in point_top:
        M_Points.append(i.text)
    points = soup_M1.find_all('td',class_='table-body__cell u-center-text')
    for i in range(1,38,2):
        M_Points.append(points[i].text)
    
    M_Rating = []
    top_rating = soup_M1.find_all('td',class_='rankings-block__banner--rating u-text-right')
    for i in top_rating:
        M_Rating.append(i.text.replace('\n','').replace('121 ...',''))
    rating = soup_M1.find_all('td',class_='table-body__cell u-text-right rating')
    for i in rating:
        M_Rating.append(i.text)
                     
    ODI_Men_Team = pd.DataFrame()
                     
    ODI_Men_Team['M_Team'] = M_Team[:10]
    ODI_Men_Team['M_Matches'] = M_Matches[:10]
    ODI_Men_Team['M_Points'] = M_Points[:10]
    ODI_Men_Team['M_Rating'] = M_Rating[:10]
    return(ODI_Men_Team)
```

```python
def batting(url):
    resource_M2 = requests.get(url)
    soup_M2 = BeautifulSoup(resource_M2.content,'html.parser')
    
    Batsmen = []
    #top batsmen name
    B_men_top = soup_M2.find_all('div',class_='rankings-block__banner--name-large')
    for i in B_men_top:
        Batsmen.append(i.text.replace('\n',''))
    #below ranked batsmen name
    B_men2 = soup_M2.find_all('td',class_='table-body__cell rankings-table__name name')
    for i in B_men2:
        Batsmen.append(i.text.replace('\n',''))
        
    Batsmen_Team = []
    #top batsmen team
    B_team_top = soup_M2.find_all('div',class_='rankings-block__banner--nationality')
    for i in B_team_top:
        Batsmen_Team.append(i.text.replace('\n',''))             
    #below ranked batsmen team
    B_team2 = soup_M2.find_all('span',class_='table-body__logo-text')
    for i in B_team2:
        Batsmen_Team.append(i.text.replace('\n',''))
    
    Batsmen_Rating = []
    #top batsmen ranking
    B_rating_top = soup_M2.find_all('div',class_='rankings-block__banner--rating')
    for i in B_rating_top:
        Batsmen_Rating.append(i.text)
    #below ranked batsmen rating
    B_rating2 = soup_M2.find_all('td',class_='table-body__cell rating')
    for i in B_rating2:
        Batsmen_Rating.append(i.text)
        
    ODI_TOP10_Batsmen = pd.DataFrame({})
    
    ODI_TOP10_Batsmen['Batsmen'] = Batsmen[:10]
    ODI_TOP10_Batsmen['Batsmen_Team'] = Batsmen_Team[:10]
    ODI_TOP10_Batsmen['Batsmen_Rating'] = Batsmen_Rating[:10]
    return(ODI_TOP10_Batsmen)
```

### 7.Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating

```python
def Amazon_mobiles(url):
    resource_7 = requests.get(url)
    soup_7 = BeautifulSoup(resource_7.text,'html.parser')
    
    Products = []
    product_name = soup_7.find_all('span',class_='a-size-base-plus a-color-base a-text-normal')
    for i in product_name:
        Products.append(i.text)
    #print(Products)   
    Price = []
    product_price = soup_7.find_all('span',class_='a-price-whole')
    for i in product_price:
        Price.append(i.text)
    #print(Price)
    Image_URL = []
    Imagelink = soup_7.find_all('div',class_='a-section aok-relative s-image-square-aspect')
    for i in Imagelink:
        Image_URL.append(i.find('img').attrs['src'])
    
    Average_Rating = []
    Ratings = []
    ratings = soup_7.find_all('span',class_='a-icon-alt')
    for i in ratings:
        Ratings.append(i.text)
    Ratings = Ratings
    for j in Ratings:
        Average_Rating.append(j[0:3])

    import pandas as pd       
    Amazon_Mobile_Products = pd.DataFrame({})

    Amazon_Mobile_Products['Products'] = Products[:10]
    Amazon_Mobile_Products['Price'] = Price[:10]
    Amazon_Mobile_Products['Image_URL'] = Image_URL[:10]
    Amazon_Mobile_Products['Average_Rating'] = Average_Rating[:10]
    return(Amazon_Mobile_Products)


    
```

### 8.Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco.

 You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
def weather(url):
    resource_8 = requests.get(url)
    soup_8 = BeautifulSoup(resource_8.content,"html.parser")
    
    Period = []
    prd = soup_8.find_all('p',class_='period-name')
    for i in prd:
        Period.append(i.text)
    
    Short_Description = []
    short_desc = soup_8.find_all('p',class_='short-desc')
    for i in short_desc:
        Short_Description.append(i.text)
    
    Temperature = []
    temp = soup_8.find_all('p',class_='temp')
    for i in temp:
        Temperature.append(i.text)
    
    Description = []
    description=soup_8.find_all('div',class_='col-sm-10 forecast-text')
    for i in description:
        Description.append(i.text)
    Description=Description[:9]
    
    Weather_Condition = pd.DataFrame({})

    Weather_Condition['Period'] = Period
    Weather_Condition['Short_Description'] = Short_Description
    Weather_Condition['Temperature'] = Temperature
    Weather_Condition['Description'] = Description
    return(Weather_Condition)
```

### 9. Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, company name, CTC, and apply date

```python
def job(url):
    resource_9 = requests.get(url)
    soup_9 = BeautifulSoup(resource_9.content,'html.parser')
    
    Job_title = []
    title = soup_9.find_all('div',class_='heading_4_5 profile')
    for i in title:
        Job_title.append(i.text.replace('\n',''))
    
    Company = []
    comp = soup_9.find_all('div',class_='heading_6 company_name')
    for i in comp:
        Company.append(i.text.replace('\n',''))
    
    CTC = []
    ctc = soup_9.find_all('div',class_='item_body')
    for i in range(1,120,3):
        CTC.append(ctc[i].text.replace('\n',''))
    
    Apply_Date = []
    date = soup_9.find_all('div',class_='item_body')
    for i in range(2,120,3):
        Apply_Date.append(date[i].text)
        
    Fresher_Jobs = pd.DataFrame()
    
    Fresher_Jobs['Job_title'] = Job_title
    Fresher_Jobs['Company'] = Company
    Fresher_Jobs['CTC'] = CTC
    Fresher_Jobs['Apply_Date'] = Apply_Date
    return(Fresher_Jobs)
```

```python
job('https://internshala.com/fresher-jobs')
```

### 10. Write a python program to scrape house details. It should include house title, location, area, emi and price

```python
def house(url):
    resource_10 = requests.get(url)
    soup_10 = BeautifulSoup(resource_10.content,'html.parser')
    
    prop_name = soup_10.find_all('h2',class_ = "heading-6 font-semi-bold nb__1AShY")
    Property = []
    for i in prop_name:
        Property.append(i.text)
    
    Location = []
    loc = soup_10.find_all('div',class_='nb__2CMjv')
    for i in loc:
        Location.append(i.text)
    
    Area = []
    area = soup_10.find_all('div',class_='nb__3oNyC')
    for i in area:
        Area.append(i.text)
    
    EMI = []
    emi = soup_10.find_all('div',class_='font-semi-bold heading-6')
    for i in range(1,len(emi),3):
        EMI.append(emi[i].text)
    
    Price = []
    price = soup_10.find_all('div',class_='font-semi-bold heading-6')
    for i in range(2,len(price),3):
        Price.append(price[i].text)
    
    Property_Details = pd.DataFrame()
    Property_Details['Property'] = Property
    Property_Details['Location'] = Location
    Property_Details['Area'] = Area
    Property_Details['EMI'] = EMI
    Property_Details['Price'] = Price
    return(Property_Details)
```
